```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# ── 配置 ──────────────────────────────────────────────────────────────────────
DATA_DIR   = "results/final"          # CSV 所在目录（相对路径，可按需修改）
MODEL      = "gpt"                    # 只画 gpt 开头的文件
MATH_TYPES = ["algebra", "geometry", "number", "probability"]
LR_VALUES  = [0, 10, 20, 50]         # 横轴轮次
LR_MAP     = {0: "lr0", 10: "lr10", 20: "lr20", 50: "lr50"}

# 颜色：每种 personality 一种颜色
PERSONALITY_COLORS = {
    "high_openness":          "#4C72B0",
    "high_conscientiousness": "#DD8452",
    "high_extraversion":      "#55A868",
    "high_agreeableness":     "#C44E52",
    "high_neuroticism":       "#8172B2",
}


# ── 读取所有 GPT CSV ──────────────────────────────────────────────────────────
records = []   # list of dicts: {math_type, lr, personality, micro_f1, macro_f1}

for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            print(f"[WARNING] 找不到文件: {fpath}")
            continue
        df = pd.read_csv(fpath)
        for _, row in df.iterrows():
            records.append({
                "math_type":   math_type,
                "lr":          lr_val,
                "personality": row["personality"],
                "micro_f1":    row["micro_f1"],
                "macro_f1":    row["macro_f1"],
            })

df_all = pd.DataFrame(records)

# ── 画图（2行×4列，第一行 Micro F1，第二行 Macro F1）─────────────────────────
fig, axes = plt.subplots(2, 4, figsize=(22, 10), sharey=False)

METRICS = [
    ("micro_f1", "Micro F1"),
    ("macro_f1", "Macro F1"),
]

for row_idx, (metric, metric_label) in enumerate(METRICS):
    for col_idx, math_type in enumerate(MATH_TYPES):
        ax = axes[row_idx][col_idx]
        df_sub = df_all[df_all["math_type"] == math_type]

        for personality, color in PERSONALITY_COLORS.items():
            df_p = df_sub[df_sub["personality"] == personality].sort_values("lr")
            if df_p.empty:
                continue
            ax.plot(
                df_p["lr"],
                df_p[metric],
                color=color,
                linestyle="-",
                marker="o",
                markersize=6,
                linewidth=1.8,
                label=personality.replace("high_", "high ").replace("_", " "),
            )

        ax.set_title(f"{metric_label}\n{math_type.capitalize()}", fontsize=12, fontweight="bold")
        ax.set_xlabel("Learning Rounds (lr)", fontsize=10)
        ax.set_ylabel(metric_label, fontsize=10)
        ax.set_xticks(LR_VALUES)
        ax.grid(True, linestyle="--", alpha=0.4)

# ── 图例（统一放底部）────────────────────────────────────────────────────────
color_handles = [
    mlines.Line2D([], [], color=c, linewidth=2, marker="o",
                  label=p.replace("high_", "high ").replace("_", " "))
    for p, c in PERSONALITY_COLORS.items()
]

fig.legend(
    handles=color_handles,
    loc="lower center",
    ncol=5,
    fontsize=11,
    frameon=True,
    bbox_to_anchor=(0.5, -0.04),
)

fig.suptitle("GPT — F1 by Personality across Learning Rounds",
             fontsize=15, fontweight="bold", y=1.01)

plt.tight_layout()
#out_path = "gpt_f1_results.png"
#plt.savefig(out_path, dpi=150, bbox_inches="tight")
#print(f"图已保存至: {out_path}")
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# ── 配置 ──────────────────────────────────────────────────────────────────────
DATA_DIR   = "results/final"
MODEL      = "gpt"
MATH_TYPES = ["algebra", "geometry", "number", "probability"]
LR_VALUES  = [0, 10, 20, 50]
LR_MAP     = {0: "lr0", 10: "lr10", 20: "lr20", 50: "lr50"}

PERSONALITIES = [
    "high_openness",
    "high_conscientiousness",
    "high_extraversion",
    "high_agreeableness",
    "high_neuroticism",
]

# 颜色 = 数学类型
MATH_COLORS = {
    "algebra":     "#4C72B0",
    "geometry":    "#DD8452",
    "number":      "#55A868",
    "probability": "#C44E52",
}

# 线型 = micro/macro
LINE_STYLES = {"micro_f1": "-", "macro_f1": "--"}
MARKERS     = {"micro_f1": "o", "macro_f1": "s"}

# ── 读取数据 ──────────────────────────────────────────────────────────────────
records = []

for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            print(f"[WARNING] 找不到文件: {fpath}")
            continue
        df = pd.read_csv(fpath)
        for _, row in df.iterrows():
            records.append({
                "math_type":   math_type,
                "lr":          lr_val,
                "personality": row["personality"],
                "micro_f1":    row["micro_f1"],
                "macro_f1":    row["macro_f1"],
            })

df_all = pd.DataFrame(records)

# ── 画图：每个 personality 一张图 ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 5, figsize=(26, 5), sharey=False)

for col_idx, personality in enumerate(PERSONALITIES):
    ax = axes[col_idx]
    df_p = df_all[df_all["personality"] == personality]

    for math_type, color in MATH_COLORS.items():
        df_m = df_p[df_p["math_type"] == math_type].sort_values("lr")
        if df_m.empty:
            continue

        for metric, ls in LINE_STYLES.items():
            ax.plot(
                df_m["lr"],
                df_m[metric],
                color=color,
                linestyle=ls,
                marker=MARKERS[metric],
                markersize=5,
                linewidth=1.8,
            )

    title = personality.replace("high_", "High ").replace("_", " ").title()
    ax.set_title(title, fontsize=12, fontweight="bold")
    ax.set_xlabel("Training Rounds (lr)", fontsize=10)
    ax.set_ylabel("F1 Score" if col_idx == 0 else "", fontsize=10)
    ax.set_xticks(LR_VALUES)
    ax.grid(True, linestyle="--", alpha=0.4)

# ── 图例：颜色=数学类型，线型=micro/macro ────────────────────────────────────
color_handles = [
    mlines.Line2D([], [], color=c, linewidth=2, marker="o",
                  label=m.capitalize())
    for m, c in MATH_COLORS.items()
]
style_handles = [
    mlines.Line2D([], [], color="black", linestyle="-",  marker="o", markersize=5, label="Micro F1"),
    mlines.Line2D([], [], color="black", linestyle="--", marker="s", markersize=5, label="Macro F1"),
]

fig.legend(
    handles=color_handles + style_handles,
    loc="lower center",
    ncol=6,
    fontsize=10,
    frameon=True,
    bbox_to_anchor=(0.5, -0.1),
)

fig.suptitle("GPT — Micro & Macro F1 by Math Type across Training Rounds",
             fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd

DATA_DIR = "results/final"

renamed = []
skipped = []

for fname in os.listdir(DATA_DIR):
    if not fname.endswith(".csv"):
        continue
    fpath = os.path.join(DATA_DIR, fname)
    df = pd.read_csv(fpath)
    cols = df.columns.tolist()

    if "empty_percentage" in cols and "empty_answer_percentage" not in cols:
        df.rename(columns={"empty_percentage": "empty_answer_percentage"}, inplace=True)
        df.to_csv(fpath, index=False)
        renamed.append(fname)
    else:
        skipped.append(fname)

print(f"已重命名 {len(renamed)} 个文件:")
for f in renamed:
    print(f"  ✓ {f}")

print(f"\n无需修改 {len(skipped)} 个文件 (列名已是 empty_answer_percentage 或两列都不存在):")
for f in skipped:
    print(f"  - {f}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ── 配置 ──────────────────────────────────────────────────────────────────────
DATA_DIR   = "results/final"
MODEL      = "gpt"
MATH_TYPES = ["algebra", "geometry", "number", "probability"]
LR_VALUES  = [10, 20, 50]
LR_MAP     = {10: "lr10", 20: "lr20", 50: "lr50"}

PERSONALITIES = [
    "high_openness",
    "high_conscientiousness",
    "high_extraversion",
    "high_agreeableness",
    "high_neuroticism",
]

PERSONALITY_COLORS = {
    "high_openness":          "#4C72B0",
    "high_conscientiousness": "#DD8452",
    "high_extraversion":      "#55A868",
    "high_agreeableness":     "#C44E52",
    "high_neuroticism":       "#8172B2",
}

# ── 读取数据 ──────────────────────────────────────────────────────────────────
records = []

for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            print(f"[WARNING] 找不到文件: {fpath}")
            continue
        df = pd.read_csv(fpath)
        for _, row in df.iterrows():
            total = row["self_study"] + row["ask_teacher"]
            ratio = row["ask_teacher"] / total if total > 0 else 0
            records.append({
                "math_type":   math_type,
                "lr":          lr_val,
                "personality": row["personality"],
                "ask_ratio":   ratio,
            })

df_all = pd.DataFrame(records)

# ── 画图：每种数学类型一张图，横轴为 personality，分组柱状图按轮次 ──────────
fig, axes = plt.subplots(1, 4, figsize=(22, 6), sharey=True)

n_personalities = len(PERSONALITIES)
n_lr = len(LR_VALUES)
bar_width = 0.22
x = np.arange(n_personalities)

LR_HATCHES = {10: "", 20: "//", 50: "xx"}   # 用纹理区分轮次

for col_idx, math_type in enumerate(MATH_TYPES):
    ax = axes[col_idx]
    df_sub = df_all[df_all["math_type"] == math_type]

    for lr_idx, lr_val in enumerate(LR_VALUES):
        df_lr = df_sub[df_sub["lr"] == lr_val]
        ratios = []
        for p in PERSONALITIES:
            row = df_lr[df_lr["personality"] == p]
            ratios.append(row["ask_ratio"].values[0] if not row.empty else 0)

        offset = (lr_idx - 1) * bar_width   # 居中对齐三组
        bars = ax.bar(
            x + offset,
            ratios,
            width=bar_width,
            color=[PERSONALITY_COLORS[p] for p in PERSONALITIES],
            hatch=LR_HATCHES[lr_val],
            edgecolor="white",
            linewidth=0.6,
            label=f"lr={lr_val}",
            alpha=0.85,
        )

    ax.set_title(math_type.capitalize(), fontsize=13, fontweight="bold")
    ax.set_xlabel("Personality", fontsize=11)
    ax.set_ylabel("Ask Teacher Ratio" if col_idx == 0 else "", fontsize=11)
    ax.set_xticks(x)
    ax.set_xticklabels(
        [p.replace("high_", "").capitalize() for p in PERSONALITIES],
        rotation=20, ha="right", fontsize=9,
    )
    ax.set_ylim(0, 1.05)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda v, _: f"{v:.0%}"))
    ax.grid(axis="y", linestyle="--", alpha=0.4)

# ── 图例：颜色=personality，纹理=轮次 ────────────────────────────────────────
import matplotlib.patches as mpatches
import matplotlib.lines as mlines

color_handles = [
    mpatches.Patch(color=PERSONALITY_COLORS[p], alpha=0.85,
                   label=p.replace("high_", "High ").replace("_", " ").title())
    for p in PERSONALITIES
]
hatch_handles = [
    mpatches.Patch(facecolor="grey", hatch=LR_HATCHES[lr], edgecolor="white",
                   alpha=0.85, label=f"lr = {lr}")
    for lr in LR_VALUES
]

fig.legend(
    handles=color_handles + hatch_handles,
    loc="lower center",
    ncol=8,
    fontsize=10,
    frameon=True,
    bbox_to_anchor=(0.5, -0.1),
)

fig.suptitle("GPT — Ask Teacher Ratio by Personality & Training Rounds",
             fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ── 配置 ──────────────────────────────────────────────────────────────────────
DATA_DIR   = "results/final"
MODEL      = "gpt"
MATH_TYPES = ["algebra", "geometry", "number", "probability"]
LR_VALUES  = [10, 20, 50]
LR_MAP     = {10: "lr10", 20: "lr20", 50: "lr50"}

PERSONALITIES = [
    "high_openness",
    "high_conscientiousness",
    "high_extraversion",
    "high_agreeableness",
    "high_neuroticism",
]
LABELS = [p.replace("high_", "").capitalize() for p in PERSONALITIES]

LR_COLORS = {10: "#4C72B0", 20: "#DD8452", 50: "#55A868"}

# ── 读取数据 ──────────────────────────────────────────────────────────────────
records = []

for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            print(f"[WARNING] 找不到文件: {fpath}")
            continue
        df = pd.read_csv(fpath)
        for _, row in df.iterrows():
            total = row["self_study"] + row["ask_teacher"]
            ratio = row["ask_teacher"] / total if total > 0 else 0
            records.append({
                "math_type":   math_type,
                "lr":          lr_val,
                "personality": row["personality"],
                "ask_ratio":   ratio,
            })

df_all = pd.DataFrame(records)

# ── 雷达图设置 ────────────────────────────────────────────────────────────────
N = len(PERSONALITIES)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]   # 闭合

fig, axes = plt.subplots(1, 4, figsize=(20, 5),
                         subplot_kw=dict(polar=True))

for col_idx, math_type in enumerate(MATH_TYPES):
    ax = axes[col_idx]
    df_sub = df_all[df_all["math_type"] == math_type]

    for lr_val in LR_VALUES:
        df_lr = df_sub[df_sub["lr"] == lr_val]
        values = []
        for p in PERSONALITIES:
            row = df_lr[df_lr["personality"] == p]
            values.append(row["ask_ratio"].values[0] if not row.empty else 0)
        values += values[:1]   # 闭合

        color = LR_COLORS[lr_val]
        ax.plot(angles, values, color=color, linewidth=2, label=f"lr={lr_val}")
        ax.fill(angles, values, color=color, alpha=0.12)

    # 轴标签
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(LABELS, fontsize=10)

    # 径向轴：0~100%
    ax.set_ylim(0, 1)
    ax.set_yticks([0.25, 0.5, 0.75, 1.0])
    ax.set_yticklabels(["25%", "50%", "75%", "100%"], fontsize=7, color="grey")

    ax.set_title(math_type.capitalize(), fontsize=13,
                 fontweight="bold", pad=15)
    ax.grid(color="grey", linestyle="--", linewidth=0.5, alpha=0.5)

# ── 图例 ──────────────────────────────────────────────────────────────────────
handles = [
    plt.Line2D([0], [0], color=LR_COLORS[lr], linewidth=2, label=f"lr = {lr}")
    for lr in LR_VALUES
]
fig.legend(handles=handles, loc="lower center", ncol=3,
           fontsize=11, frameon=True, bbox_to_anchor=(0.5, -0.06))

fig.suptitle("GPT — Ask Teacher Ratio by Personality & Training Rounds",
             fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
#多个round可以算平均整理成一个图
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ── 配置 ──────────────────────────────────────────────────────────────────────
DATA_DIR   = "results/final"
MODEL      = "gpt"
MATH_TYPES = ["algebra", "geometry", "number", "probability"]
LR_MAP     = {10: "lr10", 20: "lr20", 50: "lr50"}

PERSONALITIES = [
    "high_openness",
    "high_conscientiousness",
    "high_extraversion",
    "high_agreeableness",
    "high_neuroticism",
]
LABELS = [p.replace("high_", "").capitalize() for p in PERSONALITIES]

# ── 读取数据，按数学类型分别计算跨轮次平均 ask_ratio ─────────────────────────
records = []

for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            print(f"[WARNING] 找不到文件: {fpath}")
            continue
        df = pd.read_csv(fpath)
        for _, row in df.iterrows():
            total = row["self_study"] + row["ask_teacher"]
            ratio = row["ask_teacher"] / total if total > 0 else 0
            records.append({
                "math_type":   math_type,
                "personality": row["personality"],
                "ask_ratio":   ratio,
            })

df_all = pd.DataFrame(records)
# 对轮次求平均，保留 math_type × personality
avg = df_all.groupby(["math_type", "personality"])["ask_ratio"].mean().reset_index()

MATH_COLORS = {
    "algebra":     "#4C72B0",
    "geometry":    "#DD8452",
    "number":      "#55A868",
    "probability": "#C44E52",
}

# ── 雷达图 ────────────────────────────────────────────────────────────────────
N = len(PERSONALITIES)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))

for math_type, color in MATH_COLORS.items():
    df_m = avg[avg["math_type"] == math_type]
    values = [df_m[df_m["personality"] == p]["ask_ratio"].values[0]
              if not df_m[df_m["personality"] == p].empty else 0
              for p in PERSONALITIES]
    values += values[:1]

    ax.plot(angles, values, color=color, linewidth=2.5, label=math_type.capitalize())
    ax.fill(angles, values, color=color, alpha=0.08)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(LABELS, fontsize=11)
ax.set_ylim(0, 1)
ax.set_yticks([0.25, 0.5, 0.75, 1.0])
ax.set_yticklabels(["25%", "50%", "75%", "100%"], fontsize=8, color="grey")
ax.grid(color="grey", linestyle="--", linewidth=0.5, alpha=0.5)

ax.legend(loc="upper right", bbox_to_anchor=(1.3, 1.1), fontsize=10, frameon=True)

ax.set_title("GPT — Ask Teacher Ratio by Personality & Math Type\n(Averaged across Training Rounds)",
             fontsize=12, fontweight="bold", pad=20)

plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ── 配置 ──────────────────────────────────────────────────────────────────────
DATA_DIR   = "results/final"
MODEL      = "gpt"
MATH_TYPES = ["algebra", "geometry", "number", "probability"]
LR_VALUES  = [10, 20, 50]
LR_MAP     = {10: "lr10", 20: "lr20", 50: "lr50"}

PERSONALITIES = [
    "high_openness",
    "high_conscientiousness",
    "high_extraversion",
    "high_agreeableness",
    "high_neuroticism",
]
P_LABELS = [p.replace("high_", "").capitalize() for p in PERSONALITIES]

# 三段颜色
COLOR_SELF  = "#4C72B0"   # 自学
COLOR_ASK   = "#DD8452"   # 问老师
COLOR_EXAM  = "#55A868"   # 考试

# ── 读取数据 ──────────────────────────────────────────────────────────────────
records = []

for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            print(f"[WARNING] 找不到文件: {fpath}")
            continue
        df = pd.read_csv(fpath)
        for _, row in df.iterrows():
            records.append({
                "math_type":   math_type,
                "lr":          lr_val,
                "personality": row["personality"],
                "self_time":   row["self_study"] * 2,
                "ask_time":    row["ask_teacher"] * 3,
                "exam_time":   row["exam_time"],
            })

df_all = pd.DataFrame(records)

# ── 对所有数学类型求平均 ──────────────────────────────────────────────────────
df_avg = df_all.groupby(["lr", "personality"])[["self_time", "ask_time", "exam_time"]].mean().reset_index()

# ── 画图：单张图 ──────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(12, 6))

bar_width = 0.22
x = np.arange(len(PERSONALITIES))

for lr_idx, lr_val in enumerate(LR_VALUES):
    df_lr = df_avg[df_avg["lr"] == lr_val]
    offset = (lr_idx - 1) * bar_width

    self_vals, ask_vals, exam_vals = [], [], []
    for p in PERSONALITIES:
        row = df_lr[df_lr["personality"] == p]
        if row.empty:
            self_vals.append(0); ask_vals.append(0); exam_vals.append(0)
        else:
            self_vals.append(row["self_time"].values[0])
            ask_vals.append(row["ask_time"].values[0])
            exam_vals.append(row["exam_time"].values[0])

    positions = x + offset
    ax.bar(positions, self_vals, width=bar_width,
           color=COLOR_SELF, label="Self Study" if lr_idx == 0 else "_",
           edgecolor="white", linewidth=0.5)
    ax.bar(positions, ask_vals, width=bar_width, bottom=self_vals,
           color=COLOR_ASK, label="Ask Teacher" if lr_idx == 0 else "_",
           edgecolor="white", linewidth=0.5)
    bottom2 = [s + a for s, a in zip(self_vals, ask_vals)]
    ax.bar(positions, exam_vals, width=bar_width, bottom=bottom2,
           color=COLOR_EXAM, label="Exam" if lr_idx == 0 else "_",
           edgecolor="white", linewidth=0.5)

    # 柱顶标注轮次
    for pos, s, a, e in zip(positions, self_vals, ask_vals, exam_vals):
        total = s + a + e
        if total > 0:
            ax.text(pos, total + 1, f"lr{lr_val}", ha="center",
                    va="bottom", fontsize=7.5, color="dimgrey")

ax.set_xlabel("Personality", fontsize=11)
ax.set_ylabel("Time (timestamps, avg across math types)", fontsize=11)
ax.set_xticks(x)
ax.set_xticklabels(P_LABELS, fontsize=11)
ax.grid(axis="y", linestyle="--", alpha=0.4)

# ── 图例 ──────────────────────────────────────────────────────────────────────
import matplotlib.patches as mpatches

segment_handles = [
    mpatches.Patch(color=COLOR_SELF, label="Self Study (×2 timestamps)"),
    mpatches.Patch(color=COLOR_ASK,  label="Ask Teacher (×3 timestamps)"),
    mpatches.Patch(color=COLOR_EXAM, label="Exam"),
]
ax.legend(handles=segment_handles, loc="upper left", fontsize=10, frameon=True)

ax.set_title("GPT — Average Time Distribution by Personality & Training Rounds\n(Averaged across All Math Types)",
             fontsize=13, fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ── 配置 ──────────────────────────────────────────────────────────────────────
DATA_DIR   = "results/final"
MODEL      = "gpt"
MATH_TYPES = ["algebra", "geometry", "number", "probability"]
LR_MAP     = {0: "lr0", 10: "lr10", 20: "lr20", 50: "lr50"}

PERSONALITIES = [
    "high_openness",
    "high_conscientiousness",
    "high_extraversion",
    "high_agreeableness",
    "high_neuroticism",
]

# ── 读取数据，在每个CSV内按 macro_f1 排名 ────────────────────────────────────
records = []

for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            print(f"[WARNING] 找不到文件: {fpath}")
            continue
        df = pd.read_csv(fpath)
        # 在同一文件内按 macro_f1 降序排名（1=最高）
        df["rank"] = df["macro_f1"].rank(ascending=False, method="min")
        for _, row in df.iterrows():
            records.append({
                "math_type":   math_type,
                "lr":          lr_val,
                "personality": row["personality"],
                "rank":        row["rank"],
            })

df_all = pd.DataFrame(records)

# 对所有轮次求平均排名
avg_rank = (df_all.groupby(["personality", "math_type"])["rank"]
            .mean()
            .reset_index()
            .pivot(index="personality", columns="math_type", values="rank"))

# 调整行列顺序
avg_rank = avg_rank.loc[PERSONALITIES, MATH_TYPES]
avg_rank.index = [p.replace("high_", "High ").replace("_", " ").title()
                  for p in avg_rank.index]
avg_rank.columns = [c.capitalize() for c in avg_rank.columns]

# ── 画热力图 ──────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7, 5))

sns.heatmap(
    avg_rank,
    ax=ax,
    annot=True,
    fmt=".2f",
    cmap="RdYlGn_r",      # 绿=排名低（好），红=排名高（差）
    linewidths=0.5,
    linecolor="white",
    vmin=1, vmax=5,
    cbar_kws={"label": "Average Rank (1 = best)"},
)

ax.set_title("GPT — Average Rank by Personality & Math Type\n(Based on Macro F1, averaged across all rounds)",
             fontsize=12, fontweight="bold")
ax.set_xlabel("Math Type", fontsize=11)
ax.set_ylabel("Personality", fontsize=11)
ax.tick_params(axis="x", rotation=0)
ax.tick_params(axis="y", rotation=0)

plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import matplotlib.gridspec as gridspec
import seaborn as sns

# ══════════════════════════════════════════════════════════════════════════════
# 配置
# ══════════════════════════════════════════════════════════════════════════════
DATA_DIR   = "results/final"
MODEL      = "gpt"
MATH_TYPES = ["algebra", "geometry", "number", "probability"]
LR_ALL     = [0, 10, 20, 50]
LR_TRAIN   = [10, 20, 50]
LR_MAP_ALL = {0: "lr0", 10: "lr10", 20: "lr20", 50: "lr50"}

PERSONALITIES = [
    "high_openness",
    "high_conscientiousness",
    "high_extraversion",
    "high_agreeableness",
    "high_neuroticism",
]
P_LABELS = [p.replace("high_", "").capitalize() for p in PERSONALITIES]

P_COLORS = {
    "high_openness":          "#4C72B0",
    "high_conscientiousness": "#DD8452",
    "high_extraversion":      "#55A868",
    "high_agreeableness":     "#C44E52",
    "high_neuroticism":       "#8172B2",
}
M_COLORS = {
    "algebra":     "#17BECF",
    "geometry":    "#E377C2",
    "number":      "#BCBD22",
    "probability": "#7F7F7F",
}

FONT_TITLE  = 22
FONT_AXIS   = 20
FONT_TICK   = 18
FONT_LEGEND = 18

# ══════════════════════════════════════════════════════════════════════════════
# 读取数据
# ══════════════════════════════════════════════════════════════════════════════
records_all = []
records_tr  = []

for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP_ALL.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            print(f"[WARNING] 找不到: {fpath}")
            continue
        df = pd.read_csv(fpath)
        if "empty_percentage" in df.columns and "empty_answer_percentage" not in df.columns:
            df.rename(columns={"empty_percentage": "empty_answer_percentage"}, inplace=True)
        for _, row in df.iterrows():
            records_all.append({
                "math_type":   math_type,
                "lr":          lr_val,
                "personality": row["personality"],
                "macro_f1":    row["macro_f1"],
                "empty_pct":   row.get("empty_answer_percentage", np.nan),
            })
            if lr_val in LR_TRAIN and "self_study" in df.columns:
                total = row["self_study"] + row["ask_teacher"]
                records_tr.append({
                    "math_type":   math_type,
                    "lr":          lr_val,
                    "personality": row["personality"],
                    "ask_ratio":   row["ask_teacher"] / total if total > 0 else 0,
                    "self_time":   row["self_study"] * 2,
                    "ask_time":    row["ask_teacher"] * 3,
                    "exam_time":   row["exam_time"],
                })

df_all = pd.DataFrame(records_all)
df_tr  = pd.DataFrame(records_tr)

# ══════════════════════════════════════════════════════════════════════════════
# 图例句柄
# ══════════════════════════════════════════════════════════════════════════════
p_handles = [mlines.Line2D([], [], color=P_COLORS[p], linewidth=2.5,
                            marker="o", markersize=9, label=lbl)
             for p, lbl in zip(PERSONALITIES, P_LABELS)]
m_handles = [mlines.Line2D([], [], color=M_COLORS[m], linewidth=2.5,
                            marker="o", markersize=9, label=m.capitalize())
             for m in MATH_TYPES]

def style_ax(ax, xlabel, ylabel, xticks, title):
    ax.set_title(title, fontsize=FONT_TITLE, fontweight="bold", pad=6)
    ax.set_xlabel(xlabel, fontsize=FONT_AXIS)
    ax.set_ylabel(ylabel, fontsize=FONT_AXIS)
    ax.set_xticks(xticks)
    ax.tick_params(labelsize=FONT_TICK)
    ax.grid(True, linestyle="--", alpha=0.4)

# ══════════════════════════════════════════════════════════════════════════════
# 大图布局：4行正式图，人格图例用 fig.legend 放在 suptitle 正下方
# ══════════════════════════════════════════════════════════════════════════════
fig = plt.figure(figsize=(36, 37))
fig.patch.set_facecolor("white")

# 只用4行，间距压紧
gs0 = gridspec.GridSpec(4, 1, figure=fig, hspace=0.30,
                         top=0.91, bottom=0.04)  # top留空给标题+图例

gs_r0 = gridspec.GridSpecFromSubplotSpec(1, 4, subplot_spec=gs0[0], wspace=0.25)
gs_r1 = gridspec.GridSpecFromSubplotSpec(1, 4, subplot_spec=gs0[1], wspace=0.25)
gs_r2 = gridspec.GridSpecFromSubplotSpec(1, 5, subplot_spec=gs0[2], wspace=0.25)
gs_r3 = gridspec.GridSpecFromSubplotSpec(1, 3, subplot_spec=gs0[3], wspace=0.30,
                                          width_ratios=[1, 1.6, 1.2])

# 总标题
fig.suptitle("GPT Agent — Performance & Behavior Analysis Dashboard",
             fontsize=28, fontweight="bold", y=0.975)

# 人格图例：紧贴标题下方（y=0.955，标题y=0.985，两者相差0.03）
fig.legend(handles=p_handles,
           loc="upper center",
           bbox_to_anchor=(0.5, 0.955),
           ncol=5, fontsize=FONT_LEGEND, frameon=True,
           title="Personality", title_fontsize=FONT_LEGEND)

# ── 行0：Macro F1 × 数学类型（颜色=人格）─────────────────────────────────────
for ci, math_type in enumerate(MATH_TYPES):
    ax = fig.add_subplot(gs_r0[ci])
    df_sub = df_all[df_all["math_type"] == math_type]
    for p in PERSONALITIES:
        df_p = df_sub[df_sub["personality"] == p].sort_values("lr")
        ax.plot(df_p["lr"], df_p["macro_f1"],
                color=P_COLORS[p], linewidth=2.5, marker="o", markersize=7)
    style_ax(ax, "Learning Rounds (lr)",
             "F1" if ci == 0 else "", LR_ALL, f"F1 — {math_type.capitalize()}")

# ── 行1：Empty Answer % × 数学类型（颜色=人格）───────────────────────────────
for ci, math_type in enumerate(MATH_TYPES):
    ax = fig.add_subplot(gs_r1[ci])
    df_sub = df_all[df_all["math_type"] == math_type]
    for p in PERSONALITIES:
        df_p = df_sub[df_sub["personality"] == p].sort_values("lr")
        ax.plot(df_p["lr"], df_p["empty_pct"],
                color=P_COLORS[p], linewidth=2.5, marker="o", markersize=7)
    style_ax(ax, "Learning Rounds (lr)",
             "Empty Answer (%)" if ci == 0 else "", LR_ALL, f"Empty Answer % — {math_type.capitalize()}")

# ── 行2：Macro F1 × 人格（颜色=数学类型）────────────────────────────────────
for ci, p in enumerate(PERSONALITIES):
    ax = fig.add_subplot(gs_r2[ci])
    df_sub = df_all[df_all["personality"] == p]
    for math_type in MATH_TYPES:
        df_m = df_sub[df_sub["math_type"] == math_type].sort_values("lr")
        ax.plot(df_m["lr"], df_m["macro_f1"],
                color=M_COLORS[math_type], linewidth=2.5, marker="o", markersize=7)
    p_label = p.replace("high_", "High ").replace("_", " ").title()
    style_ax(ax, "Learning Rounds (lr)",
             "F1" if ci == 0 else "", LR_ALL, f"F1 — {p_label}")

# ── 行3左：雷达图 ─────────────────────────────────────────────────────────────
ax_radar = fig.add_subplot(gs_r3[0], polar=True)
avg_radar = df_tr.groupby(["math_type", "personality"])["ask_ratio"].mean().reset_index()
N = len(PERSONALITIES)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

for math_type, color in M_COLORS.items():
    df_m = avg_radar[avg_radar["math_type"] == math_type]
    vals = [df_m[df_m["personality"] == p]["ask_ratio"].values[0]
            if not df_m[df_m["personality"] == p].empty else 0
            for p in PERSONALITIES]
    vals += vals[:1]
    ax_radar.plot(angles, vals, color=color, linewidth=2.5, label=math_type.capitalize())
    ax_radar.fill(angles, vals, color=color, alpha=0.08)

ax_radar.set_xticks(angles[:-1])
ax_radar.set_xticklabels(P_LABELS, fontsize=FONT_TICK + 1)
ax_radar.set_ylim(0, 1)
ax_radar.set_yticks([0.25, 0.5, 0.75, 1.0])
ax_radar.set_yticklabels(["25%", "50%", "75%", "100%"], fontsize=FONT_TICK - 1, color="grey")
ax_radar.grid(color="grey", linestyle="--", linewidth=0.5, alpha=0.5)
ax_radar.set_title("Ask Teacher Ratio\nby Math Type",
                   fontsize=FONT_TITLE, fontweight="bold", pad=18)

# 数学类型图例：用 figure坐标系，紧贴雷达图右侧
# 先画完雷达图后，用 fig.legend 精确控制位置
radar_math_legend = fig.legend(
    handles=m_handles,
    loc="lower left",
    bbox_to_anchor=(0.1, 0.2),   # 雷达图右侧，行3底部
    fontsize=FONT_LEGEND, frameon=True,
    title="Math Type", title_fontsize=FONT_LEGEND,
    ncol=1
)

# ── 行3中：堆叠柱状图 ─────────────────────────────────────────────────────────
ax_bar = fig.add_subplot(gs_r3[1])
df_avg = df_tr.groupby(["lr", "personality"])[["self_time", "ask_time", "exam_time"]].mean().reset_index()
C_SELF, C_ASK, C_EXAM = "#4C72B0", "#DD8452", "#55A868"
bar_width = 0.22
x = np.arange(len(PERSONALITIES))

for li, lr_val in enumerate(LR_TRAIN):
    df_lr = df_avg[df_avg["lr"] == lr_val]
    offset = (li - 1) * bar_width
    sv, av, ev = [], [], []
    for p in PERSONALITIES:
        row = df_lr[df_lr["personality"] == p]
        sv.append(row["self_time"].values[0]  if not row.empty else 0)
        av.append(row["ask_time"].values[0]   if not row.empty else 0)
        ev.append(row["exam_time"].values[0]  if not row.empty else 0)
    pos = x + offset
    ax_bar.bar(pos, sv, width=bar_width, color=C_SELF,
               label="Self Study" if li == 0 else "_", edgecolor="white")
    ax_bar.bar(pos, av, width=bar_width, bottom=sv, color=C_ASK,
               label="Ask Teacher" if li == 0 else "_", edgecolor="white")
    b2 = [s + a for s, a in zip(sv, av)]
    ax_bar.bar(pos, ev, width=bar_width, bottom=b2, color=C_EXAM,
               label="Exam" if li == 0 else "_", edgecolor="white")
    for p2, s, a, e in zip(pos, sv, av, ev):
        ax_bar.text(p2, s + a + e + 2, f"lr{lr_val}",
                    ha="center", va="bottom", fontsize=11, color="dimgrey")

ax_bar.set_xticks(x)
ax_bar.set_xticklabels(P_LABELS, fontsize=FONT_TICK)
ax_bar.set_xlabel("Personality", fontsize=FONT_AXIS)
ax_bar.set_ylabel("Time (timestamps)", fontsize=FONT_AXIS)
ax_bar.tick_params(labelsize=FONT_TICK)
ax_bar.grid(axis="y", linestyle="--", alpha=0.4)
ax_bar.set_title("Time Distribution\nby Personality & Rounds",
                 fontsize=FONT_TITLE, fontweight="bold")
ax_bar.legend(handles=[
    mpatches.Patch(color=C_SELF, label="Self Study"),
    mpatches.Patch(color=C_ASK,  label="Ask Teacher"),
    mpatches.Patch(color=C_EXAM, label="Exam"),
], fontsize=FONT_LEGEND, loc="upper left", frameon=True)

# ── 行3右：热力图 ─────────────────────────────────────────────────────────────
ax_heat = fig.add_subplot(gs_r3[2])
rank_records = []
for math_type in MATH_TYPES:
    for lr_val, lr_tag in LR_MAP_ALL.items():
        fname = f"{MODEL}_{math_type}_{lr_tag}.csv"
        fpath = os.path.join(DATA_DIR, fname)
        if not os.path.exists(fpath):
            continue
        df = pd.read_csv(fpath)
        df["rank"] = df["macro_f1"].rank(ascending=False, method="min")
        for _, row in df.iterrows():
            rank_records.append({"math_type": math_type,
                                  "personality": row["personality"],
                                  "rank": row["rank"]})
 
df_rank = pd.DataFrame(rank_records)
pivot = (df_rank.groupby(["personality", "math_type"])["rank"]
         .mean().reset_index()
         .pivot(index="personality", columns="math_type", values="rank"))
pivot = pivot.loc[PERSONALITIES, MATH_TYPES]
pivot.index   = [p.replace("high_", "").replace("_", " ").capitalize() for p in pivot.index]
pivot.columns = [c.capitalize() for c in pivot.columns]
 
sns.heatmap(pivot, ax=ax_heat, annot=True, fmt=".2f",
            cmap="RdYlGn_r", linewidths=0.5, linecolor="white",
            vmin=1, vmax=5, annot_kws={"size": FONT_TICK + 3},
            cbar_kws={"label": "Avg Rank (1=best)", "shrink": 0.85})
ax_heat.set_title("Average Rank\nby Personality & Math Type",
                  fontsize=FONT_TITLE, fontweight="bold")
ax_heat.set_xlabel("Math Type", fontsize=FONT_AXIS)
ax_heat.set_ylabel("", fontsize=FONT_AXIS)
ax_heat.tick_params(axis="x", labelsize=FONT_TICK, rotation=0)
ax_heat.tick_params(axis="y", labelsize=FONT_TICK, rotation=0)
ax_heat.figure.axes[-1].tick_params(labelsize=FONT_TICK)
ax_heat.figure.axes[-1].set_ylabel("Avg Rank (1=best)", fontsize=FONT_AXIS)

plt.savefig("gpt_dashboard.png", dpi=150, bbox_inches="tight", facecolor="white")
print("图已保存至 gpt_dashboard.png")
plt.show()
```
